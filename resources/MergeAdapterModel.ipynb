```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import huggingface_hub

base_model_id = "mistralai/Mistral-7B-Instruct-v0.2"
peft_model_id = "torVik/mistral7binst2-bglawins"
tokenizer = AutoTokenizer.from_pretrained(peft_model_id)

base_model = AutoModelForCausalLM.from_pretrained(base_model_id, trust_remote_code=True, device_map = "auto")
model = PeftModel.from_pretrained(base_model, peft_model_id)
# del base_model

model_merged = model.merge_and_unload()

model_merged.save_pretrained("model_merged")
tokenizer.save_pretrained("model_merged")

model_merged.push_to_hub(f"{peft_model_id}_merged")
tokenizer.push_to_hub(f"{peft_model_id}_merged")
```
